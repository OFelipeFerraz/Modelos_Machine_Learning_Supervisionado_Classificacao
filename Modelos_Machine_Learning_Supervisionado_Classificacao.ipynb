# **PRÉ-PROCESSAMENTO**

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('heart.csv', sep=',', encoding='utf-8')
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

## **ATRIBUTOS PREVISORES E ALVO**

```python
df2 = pd.DataFrame.copy(df)
```

```python
df2.head()
```

```python
previsores = df2.iloc[:, 0:11].values
```

```python
previsores
```

```python
previsores.shape
```

```python
alvo = df2.iloc[:, 11].values
```

```python
print(alvo[:2])
```

```python
alvo.shape
```

## **Codificação de variáveis categóricas**

### **LabelEncoder: transformação de variáveis categóricas em numéricas**


```python
from sklearn.preprocessing import LabelEncoder
```

```python
df.head()
```

```python
previsores[:,1] = LabelEncoder().fit_transform(previsores[:,1])
previsores[:,2] = LabelEncoder().fit_transform(previsores[:,2])
previsores[:,6] = LabelEncoder().fit_transform(previsores[:,6])
previsores[:,8] = LabelEncoder().fit_transform(previsores[:,8])
previsores[:,10] = LabelEncoder().fit_transform(previsores[:,10])
```

```python
print(previsores[:2])
```

```python
previsores.shape
```

## **BASE DE TREINO E TESTE**

```python
from sklearn.model_selection import train_test_split
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(previsores, alvo, test_size = 0.3, random_state = 0)
```

```python
x_treino.shape
```

```python
x_teste.shape
```

```python
y_treino.shape
```

```python
y_teste.shape
```

# **REGRESSÃO LOGÍSTICA**

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistica = LogisticRegression(random_state=1, max_iter=1000, penalty="l2",
                               tol=0.0001, C=1,solver="lbfgs")
logistica.fit(x_treino, y_treino)
```

```python
logistica.intercept_
```

```python
logistica.coef_
```

```python
previsoes_logistica = logistica.predict(x_teste)
previsoes_logistica
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print("Acurácia: %.2f%%" % (accuracy_score(y_teste, previsoes_logistica) * 100.0))
```

```python
confusion_matrix(y_teste, previsoes_logistica)
```

```python
print(classification_report(y_teste, previsoes_logistica))
```

**Análise dados de treino**

```python
previsoes_treino = logistica.predict(x_treino)
print(previsoes_treino[:2])
```

```python
accuracy_score(y_treino, previsoes_treino)
```

```python
confusion_matrix(y_treino, previsoes_treino)
```

### **Validação Cruzada**

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
# Separando os dados em folds
kfold = KFold(n_splits = 30, shuffle=True, random_state = 5)
```

```python
# Criando o modelo
modelo = LogisticRegression(random_state=1, max_iter=1000, penalty="l2",
                               tol=0.0001, C=1,solver="lbfgs")
resultado = cross_val_score(modelo, previsores, alvo, cv = kfold)

# Usamos a média e o desvio padrão
print("Acurácia Média: %.2f%%" % (resultado.mean() * 100.0))
```

# **XGBOOST**

```python
from xgboost import XGBClassifier
```

```python
xg = XGBClassifier(max_depth=2, learning_rate=0.05, n_estimators=250, objective='binary:logistic', random_state=3)
xg.fit(x_treino,y_treino)
```

```python
previsoes_xg = xg.predict(x_teste)
previsoes_xg
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print("Acurácia: %.2f%%" % (accuracy_score(y_teste, previsoes_xg) * 100.0))
```

```python
confusion_matrix(y_teste, previsoes_xg)
```

```python
print(classification_report(y_teste, previsoes_xg))
```

**Análise dados de treino**

```python
previsoes_treino = xg.predict(x_treino)
print(previsoes_treino[:2])
```

```python
accuracy_score(y_treino, previsoes_treino)
```

```python
confusion_matrix(y_treino, previsoes_treino)
```

### **Validação Cruzada**

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
# Separando os dados em folds
kfold = KFold(n_splits = 30, shuffle=True, random_state = 5)
```

```python
# Criando o modelo
modelo = XGBClassifier(max_depth=2, learning_rate=0.05, n_estimators=250, objective='binary:logistic', random_state=3)
resultado = cross_val_score(modelo, previsores, alvo, cv = kfold)

# Usamos a média e o desvio padrão
print("Acurácia Média: %.2f%%" % (resultado.mean() * 100.0))
```

**RESULTADOS**

**REGRESSÃO LOGÍSTICA =** 83,33% e 85,27% (validação cruzada)

**XGBOOST =** 84,78% e 86,46% (validação cruzada) 

**CONCLUSÃO**

**Desempenho Geral:**

Ambos os modelos, Regressão Logística e XGBoost, demonstraram um desempenho robusto na tarefa de classificação.
A acurácia alcançada por ambos os modelos durante a validação cruzada sugere que eles são capazes de fazer previsões precisas e consistentes.


**Diferença de Desempenho:**

Observa-se que o XGBoost apresentou uma ligeira vantagem em termos de acurácia em comparação com a Regressão Logística.
Os resultados indicam que o XGBoost teve um desempenho superior, embora a diferença não seja significativa.


**Escolha do Modelo:**

A escolha entre os dois modelos pode depender de outros fatores, como interpretabilidade do modelo, tempo de treinamento e requisitos específicos do problema.
Se a interpretabilidade é crucial e o tempo de treinamento é um fator importante, a Regressão Logística pode ser uma escolha mais simples e eficiente.
Se o desempenho máximo é o principal objetivo, o XGBoost pode ser preferível, mesmo que possa exigir mais recursos computacionais.


**Otimização e Exploração Futura:**

Recomenda-se explorar opções de ajuste fino de hiperparâmetros para ambos os modelos, visando otimizar ainda mais o desempenho.
Outras métricas de avaliação, como precisão, recall e F1-score, também podem fornecer insights adicionais sobre o desempenho em diferentes aspectos da classificação.
Em resumo, os resultados sugerem que ambos os modelos são viáveis, mas a escolha entre eles dependerá das características específicas do problema e das preferências do usuário.
